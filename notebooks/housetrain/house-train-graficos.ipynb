## Univarate Data Analysis

Vamos testar alguns comandos para fazer analise univaridada simples


### Sumario Estatistico para Dados Numerico


```python
df.describe().T
```

```python
df['price'].mean()
```

```python
df.mean()
```

```python
# Calculo de quantiles
p05 = df['price'].quantile(0.05)
p25 = df['price'].quantile(0.25)
p50 = df['price'].quantile(0.50)
p75 = df['price'].quantile(0.75)
p95 = df['price'].quantile(0.95)

print(p05, p25, p50, p75, p95)
```

```python
# Medias de Price
mean = df['price'].mean()

# Desvio Padrao
std = df['price'].std()

# Minimo
minimo = df['price'].min()

# Q1
q1 = df['price'].quantile(0.25)

# Mediana
mediana = df['price'].median()

# Q2
q2 = df['price'].quantile(0.75)

# Maximo
maximo = df['price'].max()

# Skewness of Price values
skw = df['price'].skew()

# Kurtosis of Price values
kurtosis = df['price'].kurt()

# imprime os resultados
print(" Media: ", mean)
print(" Desvio Padrao: ", std)
print(" Minimo: ", minimo)
print(" Quartil 1: ", q1)
print(" Mediana: ", mediana)
print(" Quartil 2: ", q2)
print(" Maximo: ", maximo)
print(" Skewness: ", skw)
print(" Kurtosis: ", kurtosis)
```
